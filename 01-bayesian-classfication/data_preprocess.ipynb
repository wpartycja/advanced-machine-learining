```python
import pandas as pd
```

## Breast Cancer
source: https://archive.ics.uci.edu/dataset/15/breast+cancer+wisconsin+original

```python
breast_df = pd.read_csv('./data/breast-cancer-wisconsin.data', header=None)
breast_df
```

```python
breast_df.info()
```

```python
breast_df = breast_df[breast_df[6] != '?'].astype({6: 'int32'})
```

```python
breast_df.drop(columns=[0], inplace=True)
breast_df.replace({10: {2: 0, 4: 1}}, inplace=True)
breast_df.rename(columns={10: 'label'}, inplace=True)
breast_df
```

```python
breast_df.to_csv('./data/breast_cancer.csv', index=False)
```

## Occupancy detection dataset
source: https://archive.ics.uci.edu/dataset/357/occupancy+detection

```python
occupancy_df = pd.read_csv('./data/occupancy_detection/datatraining.txt')
occupancy_df
```

```python
occupancy_df.drop(columns=['date'], inplace=True)
occupancy_df.rename(columns={'Occupancy': 'label'}, inplace=True)
occupancy_df.to_csv('./data/occupancy_detection.csv', index=False)
```

## Climate dataset
source: https://www.openml.org/search?type=data&sort=runs&status=active&qualities.NumberOfClasses=%3D_2&id=1467

```python
climate_df = pd.read_csv('./data/climate.csv', header=None)
climate_df
```

```python
climate_df.replace({20: {1: 0, 2: 1}}, inplace=True)
climate_df.rename(columns={20: 'label'}, inplace=True)
climate_df
```
